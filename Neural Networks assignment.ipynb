1.=>Accuracy:
       The accuracy is the measurement of calculation that conforms to the correct valueor a standard.it is also a metric measure where it is used for overall models predictions.After hyperparameter tuning, the accuracy model may help in improving and also helps inlearning more insights for generalize the data.
        
2.=>Precision:
        precision refers to the closeness of two or more measurements to each other,indicating the consistency or repeatability of the measurements.I does not necessarily imply accuracy,which is the clossness of a measurement of its true value.A precise instument will give the same result over and over again under the same conditions.

3.=>Recall:
       Recall is a metric that measures how often a machine learning model coreectly identifies positive instances(true positives)from all the actual positive samples in the dataset.
       It also measures the abi;ity of the classifier to identify all actual positive instances.
      
4.=>F1-Score:
           F1-score is the harmonic mean of precision and recall,providing a balanced measure of both precision and recall.
      A good indicator of the performance of a classifier when the number of positive and negative instances is unbalanced.
      

Finally, hyperparameter tuning aims to optimize the model's performance by finding the best combination of hyperparameters.by fine_tuning the hyperparameters,the tuned model can achieve better accuracy,precision,recall and f1-score compared to the model with default hyperparameters.
