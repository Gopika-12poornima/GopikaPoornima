```python
transaction=[]
for i in df.itertuples():
    l=set(list(i))
    transaction.append([i for i in l if (str(i)!='nan' and type(i)!=int)])
```

```python
transaction
```

```python
len(transaction)
```

```python
t=TransactionEncoder()
t_arr=t.fit_transform(transaction)
```

```python
dt=pd.DataFrame(t_arr,columns=t.columns_)
dt
```

```python
res=apriori(dt,min_support=0.01,use_colnames=True)
res
```

```python
res=association_rules(res,metric='confidence', min_threshold=0.02)
res
```

```python
df['combined']=df.apply(lambda row:','.join(row.dropna()),axis=1)
```

```python
items=df['combined'].str.split(',')
```

```python
one_hot_encoded=pd.get_dummies(items.apply(pd.Series).stack()).sum(level=0)
```

```python
encoded_df=pd.concat([df,one_hot_encoded],axis=1)
```

```python
encoded_df.drop('combined',axis=1,inplace=True)
```
