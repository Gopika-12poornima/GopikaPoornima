```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
df=pd.read_csv('bank-full.csv',sep=';')
df
```

# EDA

```python
df.info()
```

```python
df1=pd.get_dummies(df,columns=['job','marital','education','contact','poutcome'])
df1
```

```python
pd.set_option("display.max.columns",None)
df1
```

```python
df1.info()
```

```python
df1['default']=np.where(df1['default'].str.contains("yes"),1,0)
df1['housing']=np.where(df1['housing'].str.contains("yes"),1,0)
df1['loan']=np.where(df1['loan'].str.contains("yes"),1,0)
df1['y']=np.where(df1['y'].str.contains("yes"),1,0)
df1
```

```python
df1['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
df1=df1.replace(order)
```

```python
df1
```

```python
df1.info()
```

# Model Building

```python
import warnings
warnings .filterwarnings('ignore')
```

```python
x=pd.concat([df1.iloc[:,0:11],df1.iloc[:,12:]],axis=1)
y=df1.iloc[:,11]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

# Model Predictions

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

# Testing Model Accuracy

```python
confusion_matrix=confusion_matrix(y,y_pred)
confusion_matrix
```

```python
classifier.predict_proba(x)[:,1]
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```

```python
discrete_feature=[feature for feature in df.columns if len(df[feature].unique())<20 and feature]
print('Discrete Variables Count:{}'.format(len(discrete_feature)))
```

```python
continuous_feature=[feature  for feature in df.columns if df[feature].dtype!='0'and feature not in discrete_feature]
print('continuous Variable Count:{}'.format(len(continuous_feature)))
```

# Visualizatios

```python
for feature in continuous_feature:
    df1=df.copy()
    df1[feature].hist(bins=15)
    plt.xlabel('age')
    plt.ylabel('count')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    df2=df.copy()
    if 0 in df2[feature].unique():
        pass
    else:
        df2[feature]=np.log(df2[feature])
        df2[feature].hist(bins=15)
        plt.xlabel('Age')
        plt.ylabel('count')
        plt.title(feature)
        plt.show()
```

# Outlier detection

```python
outlier = df.copy() 
fig, axes = plt.subplots(7,1,figsize=(10,8), sharex=False, sharey=False)
sns.boxplot(x='age',data=outlier,palette='crest',ax=axes[0])
sns.boxplot(x='balance',data=outlier,palette='crest',ax=axes[1])
sns.boxplot(x='day',data=outlier,palette='crest',ax=axes[2])
sns.boxplot(x='duration',data=outlier,palette='crest',ax=axes[3])
sns.boxplot(x='campaign',data=outlier,palette='crest',ax=axes[4])
sns.boxplot(x='pdays',data=outlier,palette='crest',ax=axes[5])
sns.boxplot(x='previous',data=outlier,palette='crest',ax=axes[6])
plt.tight_layout(pad=2.0)
```

# Log transformation

```python
for feature in continuous_feature:
    df3=df.copy()
    df3[feature]=np.log(df3[feature])
    df3.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
